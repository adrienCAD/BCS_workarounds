```python
# List of pokemon
search_pokemon = ['bulbasaur', 'ivysaur', 'venusaur', 'Mario', 'charmeleon',
                     'charizard', 'squirtle', 'Aang', 'blastoise']

# Set url for API
url = "https://pokeapi.co/api/v2/pokemon/"

# Set empty lists to hold pokemon's height and weight
height = []
weight = []
found_pokemon = []

# Loop through each pokemon
for pokemon in search_pokemon:
    try:
        # Create search query, make request and store in json
        query = url + pokemon
        response = requests.get(query)
        response_json = response.json()
        
        # Try to grab the height and weight of pokemon if they are available in the Pokemon API
        height.append(response_json['height'])
        weight.append(response_json['weight'])
        found_pokemon.append(pokemon)
        print(f"{pokemon} found! Appending stats")
        
    # Handle exceptions for a pokemon that is not available in the Pokemon API
    except Exception as e:
        print(f"Error occurred when retrieving {pokemon}: skipping to next one")
        pass
```

```python
# Create DataFrame
pokemon_specs = pd.DataFrame({
    'character': found_pokemon,
    'height': height,
    'mass': weight
})
pokemon_specs
```
